```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Bitcoin
The Data is downloded from [investing.com](investing.com)

```python
# Loading Bitcoin Data 
df_BTC = pd.read_csv('/content/drive/MyDrive/NAICo/Data/Bitcoin Historical Data.csv')

df_BTC.head()
```

```python
# Removing extra colmumns
df_BTC = df_BTC.drop(['Vol.', 'High', 'Low', 'Change %'], axis=1)
df_BTC = df_BTC.rename(columns={'Price': 'Close'}) 

# Change Date dtype to datetime 
df_BTC['Date'] = pd.to_datetime(df_BTC['Date'])

# New column for Rate of Return(RoR)
df_BTC['RoR'] = (df_BTC['Close'] - df_BTC['Open']) / df_BTC['Open']


df_BTC.describe()
```

```python
# Searching fo any Null values
df_BTC.isnull().values.any()
```

```python
# Reversing the indexing 
df_BTC = df_BTC[::-1]
df_BTC.reset_index(drop=True, inplace=True)

# Pick out data from May 17, 2017 to May 16, 2022 (5 years)
print(df_BTC['Date'][0])
print(df_BTC['Date'][len(df_BTC['Date'])-1])

# Droping the last row (Date = May 17, 2022)
df_BTC = df_BTC.drop([len(df_BTC['Date'])-1], axis=0)

df_BTC.tail()
```

```python
df_BTC.info()
```

```python
df_BTC.to_csv('/content/drive/MyDrive/NAICo/Cleaned Data/Bitcoin_Data.csv', index=False)
```
